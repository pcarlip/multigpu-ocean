```julia
using Plots
```

```julia
n = [1, 2, 4]

mem_per = [54823, 44871, 23207]
mem_tot = mem_per .* n

times = [4.859, 12.4835, 9.607]
```

```julia
plot(n, mem_tot, linewidth = 2, label = "Total memory", shape = :circle, markersize = 3)
plot!(
    n,
    mem_per,
    linewidth = 2,
    label = "Memory per GPU",
    shape = :circle,
    markersize = 3,
)
plot!(legend = :bottomleft)
xlabel!("Number of GPUs")
ylabel!("Memory (MiB)")
ylims!(0, 10^5)
title!("Memory use scaling")

```

```julia
plot(n, times, linewidth = 2, legend = false, shape = :circle, markersize = 3)
xlabel!("Number of GPUs")
ylabel!("Simulation time (min)")
ylims!(0, 13)
title!("Time scaling")

```

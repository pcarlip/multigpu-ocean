```julia
using Plots
```

```julia
n = [1, 2, 4]

mem = [54823, 44871*2, 23207*4]

times = [4.859, 12.4835, 9.607]
```

```julia
plot(n, mem, linewidth = 2, legend = false)
xlabel!("Number of GPUs")
ylabel!("Memory (MiB)")
title!("Memory use scaling")
```

```julia
plot(n, times, linewidth = 2, legend = false)
xlabel!("Number of GPUs")
ylabel!("Simulation time (min)")
title!("Time scaling")
```
